```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

from sklearn.datasets import load_iris
```

```python
data = load_iris(as_frame=True)
data_df = pd.DataFrame(data.data, columns=data.feature_names)
data_df['target'] = data.target
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 6))

sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', data=data_df, hue='target', palette='deep', ax=axes[0])
axes[0].set_title('Sepal Length vs Sepal Width')

sns.scatterplot(x='petal length (cm)', y='petal width (cm)', data=data_df, hue='target', palette='deep', ax=axes[1])
axes[1].set_title('Petal Length vs Petal Width')

plt.tight_layout()
plt.show()
```
